```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[4, 1],
              [2, 3]], dtype=float)

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]

    Av = A @ v
    lv = lam * v

    print(f"\nEigenpair {i+1}")
    print("A @ v =", Av)
    print("λ * v =", lv)

    assert np.allclose(Av, lv), "Eigenvector equation failed!"

# Plot eigenvectors scaled by eigenvalues
fig, ax = plt.subplots(figsize=(6, 6))

for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]

    scaled_v = lam * v

    ax.quiver(
        0, 0,
        scaled_v[0], scaled_v[1],
        angles='xy',
        scale_units='xy',
        scale=1,
        label=f'λ={lam:.2f}'
    )

ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Eigenvectors Scaled by Eigenvalues")
ax.axhline(0, color='k', lw=0.8)
ax.axvline(0, color='k', lw=0.8)
ax.grid(True)
ax.legend()

plt.tight_layout()
plt.show()

print("4.3 passed ✓")
```

### Geometric Interpretation of Eigenvalues and Eigenvectors

A matrix can be viewed as a transformation that changes vectors by rotating, stretching, shrinking, or reflecting them.

Most vectors change both their direction and length when a matrix is applied. However, some special vectors keep the same direction after the transformation. These are called **eigenvectors**.

If **v** is an eigenvector of a matrix **A**, then:

A v = λ v

where **λ** (lambda) is the corresponding **eigenvalue**.

Geometrically, the eigenvalue tells us how the matrix transforms the eigenvector:

- If **λ > 1**, the vector is stretched (becomes longer).
- If **0 < λ < 1**, the vector is squished or compressed (becomes shorter).
- If **λ < 0**, the vector is flipped in the opposite direction and scaled by |λ|.
- If **λ = 1**, the vector remains unchanged.

For example, if a matrix transforms an eigenvector from length 2 to length 6 while keeping its direction the same, then the eigenvalue is 3 because the vector was stretched by a factor of 3.

Therefore:
- **Eigenvectors** represent the directions that remain unchanged by the transformation.
- **Eigenvalues** represent how much the matrix stretches, shrinks, or flips vectors along those directions.

They help us understand the fundamental behavior of a transformation by identifying its natural axes of stretching and compression.

```python
import numpy as np

np.random.seed(42)
X = np.random.randn(4, 3)

# Compute SVD
U, S, Vt = np.linalg.svd(X, full_matrices=False)

# Reconstruct X
X_reconstructed = U @ np.diag(S) @ Vt

# Rank-1 approximation
X_approx = S[0] * np.outer(U[:, 0], Vt[0, :])

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
```

# Relationship Between SVD and PCA

Singular Value Decomposition (SVD) factorizes a matrix X into:

X = UΣVᵀ

where:

- U contains the left singular vectors.
- Σ (Sigma) contains the singular values.
- Vᵀ contains the right singular vectors.

PCA (Principal Component Analysis) seeks directions of maximum variance in the data.

The connection is:

- The columns of V (rows of Vᵀ) are the principal component directions.
- The singular values indicate how much variance is captured along each principal component.
- Larger singular values correspond to more important directions in the data.
- PCA can be computed directly using SVD on a centered data matrix.

Dimensionality reduction is achieved by keeping only the top k singular values and corresponding singular vectors. This produces a low-rank approximation that preserves most of the important structure while reducing noise and dimensionality.

In this exercise, the rank-1 approximation keeps only the largest singular value and its associated singular vectors. It captures the single most important direction of variation in the data and provides the best rank-1 approximation of the original matrix in the least-squares sense.

### Hypothesis Test: Engineering Salaries

**Null Hypothesis (H₀):**
The mean salary of Engineering employees is equal to the overall company mean salary.

H₀: μ_engineering = μ_company

**Alternative Hypothesis (H₁):**
The mean salary of Engineering employees is different from the overall company mean salary.

H₁: μ_engineering ≠ μ_company

### Interpreting the p-value

- If p-value < 0.05, reject H₀.
  There is statistically significant evidence that Engineering salaries differ from the company average.

- If p-value ≥ 0.05, fail to reject H₀.
  There is not enough evidence to conclude that Engineering salaries differ from the company average.

### Pearson Correlation Interpretation

The Pearson correlation coefficient (r) measures the strength and direction of the linear relationship between salary and years of experience.

- r > 0: positive relationship (more experience tends to correspond to higher salary)
- r < 0: negative relationship
- r ≈ 0: little or no linear relationship

The associated p-value indicates whether the observed correlation is statistically significant.

```python
import numpy as np

y_true = np.array([3.0, 5.0, 2.5, 7.0, 4.5, 6.0, 1.5, 8.0])
y_pred = np.array([2.8, 5.2, 2.1, 7.5, 4.0, 6.3, 2.0, 7.8])

n, p = len(y_true), 2

# MAE
mae = np.mean(np.abs(y_true - y_pred))

# MSE
mse = np.mean((y_true - y_pred) ** 2)

# RMSE
rmse = np.sqrt(mse)

# R²
ss_res = np.sum((y_true - y_pred) ** 2)
ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)

r2 = 1 - (ss_res / ss_tot)

# Adjusted R²
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"MAE:        {mae:.4f}")
print(f"MSE:        {mse:.4f}")
print(f"RMSE:       {rmse:.4f}")
print(f"R²:         {r2:.4f}")
print(f"Adj. R²:    {adj_r2:.4f}")
```

```python
import numpy as np
from scipy import stats
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)

s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test against standard normal distribution N(0,1)
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')

print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# Non-stationary time series (trend + noise)
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

# ADF test
adf_result = adfuller(ts)

print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once
ts_diff = np.diff(ts)

# ADF on differenced series
adf_diff = adfuller(ts_diff)

print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
```

### Kolmogorov-Smirnov (KS) Test Interpretation

The KS test compares the empirical distribution of a sample to a reference distribution.

- Null Hypothesis (H₀): The sample comes from the specified distribution (here, N(0,1)).
- Alternative Hypothesis (H₁): The sample does not come from the specified distribution.

Expected results:

- For `s1` (generated from a normal distribution), the p-value should be relatively large, so we fail to reject H₀.
- For `s2` (generated from an exponential distribution), the p-value should be very small, so we reject H₀.

### Augmented Dickey-Fuller (ADF) Test Interpretation

The ADF test checks whether a time series is stationary.

- Null Hypothesis (H₀): The series contains a unit root and is non-stationary.
- Alternative Hypothesis (H₁): The series is stationary.

For the original series:

- The series contains a positive linear trend, so the p-value is typically greater than 0.05.
- We fail to reject H₀ and conclude that the series is non-stationary.

After first differencing:

- The trend is removed.
- The p-value typically becomes less than 0.05.
- We reject H₀ and conclude that the differenced series is stationary.

This demonstrates a common technique in time-series analysis: differencing can convert a non-stationary series into a stationary one.

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    """
    epsilon = 1e-10

    # Create bins based on expected distribution
    breakpoints = np.linspace(0, 100, bins + 1)
    breakpoints = np.percentile(expected, breakpoints)

    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts, _ = np.histogram(actual, bins=breakpoints)

    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    expected_pct = np.where(expected_pct == 0, epsilon, expected_pct)
    actual_pct = np.where(actual_pct == 0, epsilon, actual_pct)

    psi = np.sum(
        (actual_pct - expected_pct) *
        np.log(actual_pct / expected_pct)
    )

    return psi


np.random.seed(1)

train_dist = np.random.normal(50, 10, 1000)
drift_dist = np.random.normal(65, 12, 1000)

psi_value = compute_psi(train_dist, drift_dist)

print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()

ax.hist(train_dist, bins=30, density=True, alpha=0.6, label='Training Distribution')
ax.hist(drift_dist, bins=30, density=True, alpha=0.6, label='Production Distribution')

ax.legend()
ax.set_title('Distribution Shift Visualization')

plt.tight_layout()
plt.show()
```

### Model Monitoring Concepts

#### Concept Drift
Concept drift occurs when the relationship between input features and the target variable changes over time, causing the model's learned patterns to become outdated.

#### Covariate Drift
Covariate drift occurs when the distribution of input features changes over time, even if the relationship between features and the target remains the same.

#### Population Stability Index (PSI) Thresholds

PSI measures how much a feature's distribution has shifted between two datasets.

- PSI < 0.10 → No significant shift (Stable)
- 0.10 ≤ PSI < 0.20 → Minor shift (Monitor closely)
- PSI ≥ 0.20 → Major shift (Investigate and potentially retrain)

#### Retraining Trigger

One common retraining trigger is when PSI exceeds 0.20 for important features, indicating a substantial distribution shift that may degrade model performance.

### Conditional Probability

P(second=blue | first=red)

Two events A and B are independent if:

P(B|A) = P(B)

Here:

P(blue | red) = 3/9 = 0.3333

P(blue) = 3/10 = 0.3000

Since:

P(blue | red) ≠ P(blue)

### Normal Distribution
Used in Gaussian Naive Bayes and for modeling continuous features that approximately follow a bell-shaped distribution.

### Binomial Distribution
Used in binary classification problems where each trial has two possible outcomes (success/failure).

```python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (
    p_free_given_spam * p_spam
    + p_free_given_ham * p_ham
)

# Bayes' theorem
p_spam_given_free = (
    p_free_given_spam * p_spam
) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    p_ham = 1 - prior_spam

    evidence = (
        p_word_given_spam * prior_spam
        + p_word_given_ham * p_ham
    )

    posterior = (
        p_word_given_spam * prior_spam
    ) / evidence

    return posterior

pred = naive_bayes_predict(0.30, 0.80, 0.05)

assert abs(pred - p_spam_given_free) < 1e-9

print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
```

### Bayes' Theorem in the Spam Filter Example

Bayes' Theorem:

P(Spam | "free") = P("free" | Spam) × P(Spam) / P("free")

#### Prior
The probability that an email is spam before seeing any words.

P(Spam) = 0.30

#### Likelihood
The probability of observing the word "free" given that the email is spam.

P("free" | Spam) = 0.80

#### Evidence
The overall probability of observing the word "free" in any email.

P("free") = P("free" | Spam)P(Spam)
          + P("free" | Not Spam)P(Not Spam)

= 0.275

#### Posterior
The updated probability that an email is spam after observing the word "free".

P(Spam | "free") = 0.8727

This means that if an email contains the word "free", there is approximately an 87.27% probability that it is spam according to this model.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(7)

population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30

# Generate 5000 sample means
sample_means = np.array([
    np.random.choice(population, size=sample_size, replace=True).mean()
    for _ in range(n_samples)
])

pop_mean = population.mean()
pop_std  = population.std()

# CLT predicted standard deviation
clt_std = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot histogram + theoretical normal curve
fig, ax = plt.subplots()

ax.hist(
    sample_means,
    bins=40,
    density=True,
    alpha=0.7,
    label='Sample Means'
)

x = np.linspace(
    sample_means.min(),
    sample_means.max(),
    500
)

ax.plot(
    x,
    stats.norm.pdf(x, loc=pop_mean, scale=clt_std),
    linewidth=2,
    label='CLT Normal Approximation'
)

ax.legend()
ax.set_title('CLT: Distribution of Sample Means')

plt.tight_layout()
plt.show()

# KS test
ks_result = stats.kstest(
    sample_means,
    'norm',
    args=(pop_mean, clt_std)
)

print(
    f"KS test p-value: {ks_result.pvalue:.4f} "
    f"→ Approximately normal? {ks_result.pvalue > 0.05}"
)
```

### Central Limit Theorem (CLT)

The Central Limit Theorem states that if we repeatedly draw random samples of size n from any population with a finite mean and variance, the distribution of the sample means approaches a normal distribution as n becomes large.

In this experiment:

- The population follows an Exponential distribution, which is highly skewed and non-normal.
- We repeatedly drew 5,000 samples of size 30.
- The distribution of the resulting sample means appears approximately normal.
- The mean of the sample means is close to the population mean.
- The standard deviation of the sample means is close to:

σ / √n

which is exactly what the CLT predicts.

The KS test provides a quantitative check of how closely the sample means follow the predicted normal distribution.
